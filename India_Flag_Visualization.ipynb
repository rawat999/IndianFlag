```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patch
```

```python
# Added Color Palettes

# Green
a = patch.Rectangle((0, 1), width=12, height=2.66, facecolor='#138808', edgecolor='grey')

# White
b = patch.Rectangle((0, 3.66), width=12, height=2.68, facecolor='#FFFFFF', edgecolor='grey')

# Saffron
c = patch.Rectangle((0, 6.34), width=12, height=2.66, facecolor='#FF9933', edgecolor='grey')

# Add into plot
m, n = plt.subplots(figsize=(18,12))
n.add_patch(a)
n.add_patch(b)
n.add_patch(c)

# AshokChakra Circle
radius=1.33
plt.plot(6, 4.99, marker='o', markerfacecolor='#000088ff', markersize=9)
chakra = plt.Circle((6, 4.99), radius, color='#000088ff', fill=False, linewidth=7)
n.add_patch(chakra)

# 24 spokes in AshokChakra
for i in range(0, 24):
  p = 6 + radius/2 * np.cos(np.pi*i/12 + np.pi/48)
  q = 6 + radius/2 * np.cos(np.pi*i/12 - np.pi/48)
  r = 4.99 + radius/2 * np.sin(np.pi*i/12 + np.pi/48)
  s = 4.99 + radius/2 * np.sin(np.pi*i/12 - np.pi/48)
  t = 6 + radius * np.cos(np.pi*i/12)
  u = 4.99 + radius * np.sin(np.pi*i/12)
  n.add_patch(patch.Polygon([[6, 4.99], [p, r], [t, u], [q, s]], fill=True, closed=True, color='#000088ff'))
plt.axis('off')
plt.show()
```
